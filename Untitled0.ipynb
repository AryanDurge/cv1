```python
import pandas as pd
def load_csv(filepath):
  #print(filepath)
  data = []
  col = []
  checkcol = False
  with open(filepath)as f:
    #print(f)
    for val in f.readlines():
      val = val.replace("\n","")
      val = val.split(',')
      #print(val)
      if checkcol is False:
        col = val
        #print(col)
        checkcol = True
      else:
        data.append(val)
  df = pd.DataFrame(data=data,columns=col)
  return df
myData = load_csv("/content/drive/MyDrive/Datasets/california_housing_test.csv")
print(myData.head())
```

```python
#Data Exploration
#Import Pandas Package
import pandas as pd
import numpy as np

#assign data

data = {"game":['jai','princi','Gaurav',
                'Anuj','Natasha','Riya'],
        'Age':[17,20,23,12,43,23],
        'Gender':['M','F','M','M','F','F'],
        'Marks':[90,76,'NaN',74,65,71]}

#convert into DataFrame

df = pd.DataFrame(data)

print(df.head())
```

```python
# data.isnull()

c = avg = 0
for ele in df['Marks']:
  if str(ele).isnumeric():
    c += 1
    avg += ele
    avg = avg/c

df = df.replace(to_replace="NaN",value=avg)

  #Display data
df
```

```python
#Reshaping the data

df['Gender']=df['Gender'].map({'M':0 ,'F':1}).astype(float)
#display data
df
```

```python
df['Marks'] = pd.to_numeric(df['Marks'],errors="coerce")
df=df[df['Marks'] >= 75]
df=df.drop(['Age'],axis=1)
df
```

```python
#rescaling a feature
#use sckit-learn's MinMaxScaler to rescale a feature array:
#load libraries

import numpy as np
from sklearn import preprocessing

#create feature
feature = np.array([[-500.5],[-100.1],[0],[100.1],[900.9]])

#create scaler
minmax_scale = preprocessing.MinMaxScaler(feature_range=(0,1))

#scale feature
scaled_feature = minmax_scale.fit_transform(feature)

#show feature
scaled_feature
```

```python
import numpy as np
from sklearn import preprocessing

x = np.array([[-1000.1],[-200.2],[500.5],[600.6],[9000.9]])

#create scaler
scaler = preprocessing.StandardScaler()

#tranform the feature
standardized = scaler.fit_transform(x)
standardized
```

```python
#print mean and standard deviation
print('Mean',round(standardized.std()))
print("standardized devaition",standardized.std())
```

```python
import numpy as np
from sklearn.preprocessing import Normalizer

features = np.array([[0.5,0.5],
                     [1.1,3.4],
                     [1.5,3.4],
                     [1.63,34.4],
                     [10.9,3.3]])
#creating feature matrix
normalizer = Normalizer(norm="l2")

#Transformer feature matrix

normalizer.transform(features)
```

```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer
features = np.array([[2,3],
                     [2,3],
                     [2,3]])
def add_ten(x):
  return x + 10

ten_transformer = FunctionTransformer(add_ten)

ten_transformer.transform(features)

```
